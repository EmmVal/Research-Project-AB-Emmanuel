```python
import pandas as pd
import numpy as np
import os
import re
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
```

```python
# Set directories

# Directory for Raw datasets (not uploaded in Github repository)
raw_data_path = './Raw_data/'

#Directory for transformed data
transformed_data_path = './Transformed_data/'
```

```python
all_files_list = os.listdir(raw_data_path) #list with all files

full_db = pd.DataFrame() #dataset to concat all information

#Load tour-level main draw matches data and append into a single dataset
for file in all_files_list:
    if bool(re.match(r"^atp_matches_\d{4}\.csv$" , file)): #Only load tour-level main draw matches
            df = pd.read_csv(raw_data_path+file)
            #print(f'Reading file {file}, size {df.shape}') 
            full_db = pd.concat([full_db,df],axis = 0)

print('\nSize of total file', full_db.shape,'\n')
```

```python
#Create season varaible
full_db['Season'] = [int(x/10000) for x in full_db['tourney_date']]
```

```python
full_db_2000_20019 = full_db[(full_db['Season']>=2000)&(full_db['Season']<=2019)].reset_index(drop=True)
full_db_2000_20019.shape
```

```python
full_db_2000_20019['Win_indicator'] = [1 if x<y else 0 for x,y in zip(full_db_2000_20019['winner_rank'],full_db_2000_20019['loser_rank'])]
full_db_2000_20019['Win_indicator'] = np.where(full_db_2000_20019['winner_rank'].isnull(),np.nan, full_db_2000_20019['Win_indicator'])
full_db_2000_20019['Win_indicator'] = np.where(full_db_2000_20019['loser_rank'].isnull(),np.nan, full_db_2000_20019['Win_indicator'])
```

```python
#full_db_2000_20019['Diff_points'] = np.abs(full_db_2000_20019['winner_rank_points']-full_db_2000_20019['loser_rank_points'])
full_db_2000_20019['Diff_points'] = np.where(full_db_2000_20019['Win_indicator']==1, 
                                             full_db_2000_20019['winner_rank_points']-full_db_2000_20019['loser_rank_points'],
                                             full_db_2000_20019['loser_rank_points']-full_db_2000_20019['winner_rank_points'])

full_db_2000_20019['Diff_points'] = np.where(full_db_2000_20019['Win_indicator'].isnull(),np.nan, full_db_2000_20019['Diff_points'])
full_db_2000_20019['Diff_points'] = np.where(full_db_2000_20019['winner_rank_points'].isnull(),np.nan, full_db_2000_20019['Diff_points'])
full_db_2000_20019['Diff_points'] = np.where(full_db_2000_20019['loser_rank_points'].isnull(),np.nan, full_db_2000_20019['Diff_points'])
```

```python
selected = ['Win_indicator','winner_rank','loser_rank','winner_rank_points','loser_rank_points','Diff_points']
full_db_2000_20019[selected]
```

```python
Selected = ['Season','tourney_date','tourney_id','match_num','Win_indicator','winner_id','loser_id','Diff_points']
full_db_selected = full_db_2000_20019[Selected]
full_db_selected = full_db_selected.sort_values(by=['tourney_date','tourney_id','match_num']).reset_index(drop=True)
full_db_selected.shape
```

```python
# Initialize Elo dictionary (default = 1500)
elo = {}

#Define function to compute probabilities from elo scores
def pi_elo(Elo_i, Elo_j):
    return 1 / (1 + 10 ** ((Elo_j - Elo_i) / 400))

#Gets current elo for player_id
def get_elo(player_id):
    return elo.get(player_id, 1500)

#Updates Elo ranking for winner and losser after match
def update_elo(winner_id, loser_id, K = 32):
    Elo_i_bf = get_elo(winner_id)
    Elo_j_bf = get_elo(loser_id)

    pi_ij = pi_elo(Elo_i_bf, Elo_j_bf)
    pi_ji = 1 - pi_ij

    Elo_i_new = Elo_i_bf + K * (1 - pi_ij) #updates Elo for winner
    Elo_j_new = Elo_j_bf + K * (0 - pi_ji) #update Elo for losser

    #Save current Elos for player in dictionary
    elo[winner_id] = Elo_i_new
    elo[loser_id] = Elo_j_new

    return Elo_i_bf, Elo_j_bf, pi_ij, pi_ji, Elo_i_new, Elo_j_new

def elo_history_all(df):
    Selected = ['tourney_date','tourney_id','match_num', 'winner_id','loser_id']
    df_s = df[Selected]
    df_s = df_s.sort_values(by=['tourney_date','tourney_id','match_num']).reset_index(drop=True)
    
    # List to track Elo after each match
    history = []
    
    # Compute Elo for each row
    for _, row in df_s.iterrows():
        Elo_bf_winner, Elo_bf_losser, pi_winner, pi_losser, Elo_new_winner, Elo_new_losser = update_elo(row["winner_id"],
                                                                                                          row["loser_id"])
        history.append({
            "tourney_date": row["tourney_date"],
            "tourney_id": row["tourney_id"],
            "match_num": row["match_num"],
            "winner_id": row["winner_id"],
            "loser_id": row["loser_id"],
            "Elo_winner_before": Elo_bf_winner,
            "Elo_losser_before": Elo_bf_losser,
            "Pi_winner": pi_winner,
            "PI_losser": pi_losser,
            "Elo_winner_new": Elo_new_winner,
            "Elo_losser_new": Elo_new_losser
        })
    
    return pd.DataFrame(history)

elo_history_db = elo_history_all(full_db_selected)
print(elo_history_db.shape)
elo_history_db.head()
```

```python
full_db_selected = full_db_selected.merge(elo_history_db, how = 'left', on =['tourney_date','tourney_id','match_num', 'winner_id','loser_id'])
print(full_db_selected.shape)
full_db_selected.head()
```

```python
Train_db = full_db_selected_dp[full_db_selected_dp['Season']<=2018]
Test_db = full_db_selected_dp[full_db_selected_dp['Season']>2018]
```

```python
from sklearn.preprocessing import StandardScaler

X_train = np.array(Train_db['Diff_points'])
X_test = np.array(Test_db['Diff_points'])

X_train_scaled = X_train.reshape(-1,1)#StandardScaler().fit_transform(X_train.reshape(-1,1))
X_test_scaled = StandardScaler().fit_transform(X_test.reshape(-1,1))

y_train = np.array(Train_db['Win_indicator'])
y_test = np.array(Test_db['Win_indicator'])
```

#### Logistic regression

```python
# Logistic regression without intercept
log_reg = LogisticRegression(fit_intercept=False, solver="lbfgs")
log_reg.fit(X_train_scaled, y_train)
print("Coefficient:", log_reg.coef_[0][0])

y_pred_prob_train = log_reg.predict_proba(X_train_scaled)[:, 1]
y_pred_train = log_reg.predict(X_train_scaled)
y_pred_prob_test = log_reg.predict_proba(X_test_scaled)[:, 1]
y_pred_test = log_reg.predict(X_test_scaled)

#import statsmodels.api as sm
#logit_model = sm.Logit(y, X_scaled)
#result = logit_model.fit()
#print(result.summary())
```

#### Metrics

```python
def accuracy(y_truth, y_pred):
    return float(sum(y_truth==y_pred)/y_truth.shape[0])

def calibration(y_truth, y_prob):
    return float(sum(y_prob)/sum(y_truth))

def log_loss(y_truth, y_prob):
    return float(-np.mean(y_truth * np.log(y_prob) + 
                          (1 - y_truth) * np.log(1 - y_prob)))

def performance_metrics(y_truth, y_pred, y_prob):
    acc = accuracy(y_truth, y_pred)
    cal = calibration(y_truth, y_prob)
    log_ls = log_loss(y_truth, y_prob)
    return acc, cal, log_ls

def print_metrics(y_truth, y_pred, y_prob):
    acc, cal, log_ls = performance_metrics(y_truth, y_pred, y_prob)
    print(f'Accuracy: {round(acc,4)}, Calibration: {round(cal,4)}, Log_loss: {round(log_ls,4)}')

print('_____ Logistic Regression metrics _____')
print('**Train Sample**')
print_metrics(y_train, y_pred_train, y_pred_prob_train)
print('**Test Sample**')
print_metrics(y_test, y_pred_test, y_pred_prob_test)
```

#### Elo model

```python
y_pred_prob_train = np.array(Train_db['Pi_winner'])
y_pred_train = np.array(Train_db['Pi_winner']>.5)
y_pred_prob_test = np.array(Test_db['Pi_winner'])
y_pred_test = np.array(Test_db['Pi_winner']>.5)

print('_____ Logistic Regression metrics _____')
print('**Train Sample**')
print_metrics(y_train, y_pred_train, y_pred_prob_train)
print('**Test Sample**')
print_metrics(y_test, y_pred_test, y_pred_prob_test)
```
