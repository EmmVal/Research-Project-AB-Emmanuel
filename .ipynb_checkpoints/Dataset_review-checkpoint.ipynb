## Libraries and directories

```python
import pandas as pd
import numpy as np
import os
import re
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Set directories

# Directory for Raw datasets (not uploaded in Github repository)
raw_data_path = './Raw_data/'

#Directory for transformed data
transformed_data_path = './Transformed_data/'
```

## Data review

### Load data

```python
all_files_list = os.listdir(raw_data_path) #list with all files

full_db = pd.DataFrame() #dataset to concat all information

#Load tour-level main draw matches data and append into a single dataset
for file in all_files_list:
    if bool(re.match(r"^atp_matches_\d{4}\.csv$" , file)): #Only load tour-level main draw matches
            df = pd.read_csv(raw_data_path+file)
            #print(f'Reading file {file}, size {df.shape}') 
            full_db = pd.concat([full_db,df],axis = 0)

print('\nSize of total file', full_db.shape,'\n')
```

```python
full_db.head()
```

### Create season column

```python
full_db['Season'] = [int(x/10000) for x in full_db['tourney_date']]
full_db['Season'].unique()
```

### Filter data for 21st Century Golden Era.

```python
#Filter data since 1999, as this was first appeareance in Grand Slam of one of the Big-Three (Roger Federer)
full_db_ge = full_db[full_db['Season']>=1999].reset_index(drop=True)
print('\nSize of DB from Golden Era', full_db_ge.shape,'\n')
```

### Filter data only for four Grand Slams

```python
full_db_gs =  full_db_ge[full_db_ge['tourney_level']=='G'].reset_index(drop=True)
print('\nSize of DB with Grand Slams only', full_db_gs.shape,'\n')
```

### Validations

```python
#Validation for Tournament name
full_db_gs['tourney_name'].value_counts()
```

```python
# Treatment for Grand Slam name
full_db_gs['tourney_name'] = np.where(full_db_gs['tourney_name'].isin(['Roland Garros']),'French Open',
                             np.where(full_db_gs['tourney_name'].isin(['Us Open']),'US Open',
                                      full_db_gs['tourney_name']))

full_db_gs['tourney_name'].value_counts()
```

#### Create target variable

```python
full_db_gs['Win_indicator'] = np.where(full_db_gs['winner_rank']<full_db_gs['loser_rank'],1,0 )
full_db_gs['Win_indicator'] = np.where(full_db_gs['winner_rank'].isnull(),np.nan, full_db_gs['Win_indicator'])
full_db_gs['Win_indicator'] = np.where(full_db_gs['loser_rank'].isnull(),np.nan, full_db_gs['Win_indicator'])

print('Number of cases with missing in ranking/Target:',sum(full_db_gs['Win_indicator'].isnull()))
review_prop = full_db_gs.dropna(subset=['Win_indicator'])
```

```python
#Validation for 128 players in each tournament
full_db_gs['draw_size'].value_counts()
```

```python
# Validate number of matches per tournament
match_num_gp = full_db_gs.groupby(['Season','tourney_name'])['match_num'].count().reset_index()
print('Matches per tournament',match_num_gp['match_num'].unique())

#127 matches per tournament - Ok
```

## Analysis

### Championship analysis

#### Identify the match for the championship

```python
#Paste maximum match played in tournament
match_max_gp = full_db_gs.groupby(['Season','tourney_name'])['match_num'].max().reset_index()
match_max_gp = match_max_gp.rename(columns={'match_num': 'max_match'})
full_db_gs = pd.merge(full_db_gs,match_max_gp, how='left', on=['Season','tourney_name'])

#Create indicator for championship match
full_db_gs['Champion_match'] = [1 if x ==y else 0 for x, y in zip(full_db_gs['match_num'],full_db_gs['max_match'])]
full_db_gs
```

#### Extract champion winners from the Golden Era

```python
# Define Big-three players from modern era
big_three = ["Novak Djokovic", "Rafael Nadal", "Roger Federer"]
```

```python
championship_matches = full_db_gs[full_db_gs['Champion_match']==1].reset_index(drop=True)
championship_matches.shape
```

```python
#Indicator of big three
championship_matches['Big 3 Indicator'] = np.where(championship_matches['winner_name'].isin(big_three),'Big 3','Other')
championship_matches['Big 3 Indicator'].value_counts()
```

```python
champions_per_season = championship_matches.groupby(['Season','Big 3 Indicator'])['tourney_id'].count().reset_index()

def historic_champions_big_3(db):
    pivot_df = db.pivot(index="Season", columns="Big 3 Indicator", values="tourney_id").fillna(0)
    
    # Plot with custom colors
    ax = pivot_df.plot( kind="bar", stacked=True, figsize=(12,6), color=["gold", "lightseagreen"] )
    
    # Titles and labels
    ax.set_title("Grand Slam Matches Won Per Season")
    ax.set_ylabel("Number of Matches Won")
    ax.set_xlabel("Season")
    
    # Remove top and left spines
    ax.spines['top'].set_visible(False)
    ax.spines['left'].set_visible(False)
    
    # Make y-axis integers
    ax.yaxis.get_major_locator().set_params(integer=True)
    plt.show()


historic_champions_big_3(champions_per_season)
```
